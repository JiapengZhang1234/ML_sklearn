```python
from sklearn.svm import LinearSVC,SVC
import matplotlib.pyplot as plt 
from sklearn.model_selection import cross_val_score
from sklearn.datasets import make_blobs

#实例1

#数据准备
x,y = make_blobs(n_samples=50, centers=2,random_state=0,cluster_std=0.6)
plt.scatter(x[:,0],x[:,1],c=y,s=50,cmap="rainbow")
plt.show()
```

```python
#模型性能交叉验证


svc1 = cross_val_score(LinearSVC(),x,y,cv=5,scoring='accuracy')

svc2 = cross_val_score(SVC(kernel='linear'),x,y,cv=5,scoring='accuracy')


print(svc1.mean())
print(svc2.mean())
```

```python
#实例2

from sklearn.datasets import make_circles

#数据准备
x,y = make_circles(n_samples=1000, noise=0.03,factor=0.6)
plt.scatter(x[:,0],x[:,1],c=y,s=10,cmap="rainbow")
plt.show()
```

```python
#模型性能交叉验证

svc3 = cross_val_score(SVC(kernel='linear'),x,y,cv=5,scoring='accuracy')

svc4 = cross_val_score(SVC(kernel='rbf'),x,y,cv=5,scoring='accuracy')


print(svc3.mean())
print(svc4.mean())
```
